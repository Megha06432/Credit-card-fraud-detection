```python
from sklearn.preprocessing import StandardScaler, PowerTransformer
```

```python
X = a.drop([ 'Class'], axis=1)
y=a.iloc[:,-1]
```

```python


X
```

```python
if 'Time' in X.columns:
    X = X.drop('Time', axis=1)
else:
    print("'Time' column not found in the dataset. Skipping this step.")
```

```python
if 'Amount' in X.columns:
    X['Amount'] = np.log1p(X['Amount'])
else:
    print("'Amount' column not found in the dataset. Skipping log transformation.")
```

```python
# Apply PowerTransformer to make data more Gaussian-like
pt = PowerTransformer(method='yeo-johnson')
X = pd.DataFrame(pt.fit_transform(X), columns=X.columns)
```

```python
X.replace([np.inf, -np.inf], np.nan, inplace=True)
X.dropna(inplace=True)
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=7, test_size=0.2);X_train, X_test, y_train, y_test 
```

```python
import pandas as pd
skewness_values=a.skew();skewness_values
skewness_values = skewness_values.replace([float('inf'), float('-inf')], pd.NA)
# Convert infinite values to NaN
with pd.option_context('mode.use_inf_as_na', True):
    # Your code here
    sns.barplot(x=skewness_values.values, y=skewness_values.index, color='skyblue')
    plt.xlabel('Skewness')
    plt.ylabel('Feature')
    plt.title('Skewness of Features')
    plt.grid(axis='x')
    plt.tight_layout()
    plt.show()
```

```python
def evaluate_models(X_train, y_train, X_test, y_test, models):
    results = []
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        mcc = matthews_corrcoef(y_test, y_pred)
        results.append((name, accuracy, precision, recall, f1, mcc))
    return results
```
